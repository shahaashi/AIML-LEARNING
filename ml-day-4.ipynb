```python
import numpy as np
import matplotlib.pyplot as plt

X = np.array([1,2,3,4])
y = np.array([7,9,11,13])

w = 0
b = 0
learning_rate = 0.01

for i in range(1000):
    y_pred = w * X + b
    dw = (2/len(X)) * np.dot(X, (y_pred - y))
    db = (2/len(X)) * np.sum(y_pred - y)
    w -= learning_rate * dw
    b -= learning_rate * db

    error = y_pred - y
    loss = np.mean(error**2)

    if (i + 1) % 100 == 0:
        print(f"Iteration {i+1}: w={w:.4f}, b={b:.4f}, Loss={loss:.4f}")

plt.scatter(X, y, label="Actual Data")
plt.plot(X, y_pred, label="Regression Line")

plt.xlabel("X")
plt.ylabel("y")
plt.title("Linear Regression Fit")
plt.legend()

plt.show()

```
