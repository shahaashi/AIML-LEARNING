Evaluate classification model properly

confusion matrix
precision
recall
F1 score

```python
import numpy as np
actual_labels = [0,0,1,1,0,0,1,1]
predicted_labels = [0,0,1,1,0,0,1,1]

tp = np.sum((np.array(actual_labels) == 1) & (np.array(predicted_labels) == 1))
tn = np.sum((np.array(actual_labels) == 0) & (np.array(predicted_labels) == 0))
fp = np.sum((np.array(actual_labels) == 0) & (np.array(predicted_labels) == 1))
fn = np.sum((np.array(actual_labels) == 1) & (np.array(predicted_labels) == 0))
print("True Positives:", tp)
print("True Negatives:", tn)        
print("False Positives:", fp)
print("False Negatives:", fn)

precision = tp / (tp + fp) if (tp + fp) > 0 else 0
recall = tp / (tp + fn) if (tp + fn) > 0 else 0
print("Precision:", precision)
print("Recall:", recall)
f1_score = 2 * (precision * recall) / (precision + recall) if (precision + recall) > 0 else 0
print("F1 Score:", f1_score)
```
