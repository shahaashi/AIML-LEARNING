Understand Overfitting, Underfitting, and Regularization

So far you trained models on small clean datasets.
In real machine learning systems, models often fail because they either:

learn too little

or learn too much noise

This leads to two important concepts.

1. Underfitting

Underfitting happens when the model is too simple to learn the pattern.

Example:

True pattern: curved relationship
Model: straight line

Result:

high training error

high testing error

Example graph idea:

data:  curved pattern
model: straight line

The model cannot capture the relationship.

2. Overfitting

Overfitting happens when the model learns noise instead of the real pattern.

Example:

data: slightly curved pattern
model: extremely complex curve touching every point

Result:

very low training error

very high testing error

The model memorizes the dataset instead of learning the underlying rule.

3. Ideal Model

The goal is a model that captures the general pattern.

training error: low
testing error: low

Graph intuition:

underfit → simple line
good fit → smooth curve
overfit → crazy curve through all points
4. Why Overfitting Happens

Common causes:

Cause	Example
too many parameters	large neural network
small dataset	few training samples
too many features	irrelevant variables
too many training iterations	memorization

```python
import numpy as np

X = np.array([1,2,3,4,5], dtype=np.float32)
y = np.array([2,4,6,8,10], dtype=np.float32)

w = 0.0
b = 0.0
learning_rate = 0.01
lambda_reg = 0.01
for i in range(1000):
    y_pred = w * X + b
    error = y_pred - y

    loss = np.mean(error**2) + lambda_reg * (w**2)
    dw = 2 * np.mean(error * X) + 2 * lambda_reg * w
    db = 2 * np.mean(error)

    w -= learning_rate * dw
    b -= learning_rate * db

    if i % 500 == 0:
        print("Loss:", loss, "Weights:", w, "Bias:", b)
print("Final weights:", w)
print("Final bias:", b) 
print("Final loss:", loss)

```
