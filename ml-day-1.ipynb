Task 1 — Compute basic statistics

Write functions that compute:

Mean

Variance

Standard Deviation

You must implement these manually without using libraries like NumPy or statistics.

Your functions should work for any list of numbers.

```python
data = [10,20,30,40,50]

mean = sum(data) / len(data)
print("Mean:", mean)

vars = [(x - mean) ** 2 for x in data]
variance = sum(vars) / len(data)    
print("Variance:", variance)

std_dev = variance ** 0.5
print("Standard Deviation:", std_dev)       
```

Task 2 — Normalize a dataset

Implement Min–Max Normalization.

Formula:

normalized_value = (x - min_value) / (max_value - min_value)

The function should convert values into the range:

0 → 1

Example:

Input:

[10, 20, 30, 40, 50]

Output:

[0.0, 0.25, 0.5, 0.75, 1.0]

```python
def normalized_data(data):
    min_value = min(data)
    max_value = max(data)
    return [(x - min_value) / (max_value - min_value) for x in data]
print("Normalized Data:", normalized_data(data))
```
