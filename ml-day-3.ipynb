Day 3 — Linear Regression (Concept)
1. The Problem Machine Learning Tries to Solve

Suppose we have data:

House Size	Price
1000	50
1200	55
1500	65
1800	80

We want a model that predicts:

house_size → price

So we want to learn a rule like:


price=w×size+b

Where:

w → weight

b → bias

But we do not know the correct values.

The model must learn them from data.

2. Prediction Function

A model predicts using:

y^​=wx+b

Example:

w = 0.04
b = 10
x = 1200

Prediction:

0.04 * 1200 + 10 = 58
3. The Error (Loss)

Predictions will not match the real values.

Example:

Actual Price	Predicted Price
55	58

Error:

error = predicted - actual

To measure how wrong the model is we use a loss function.

Common one:

Mean Squared Error
MSE=n1​∑(y−y^​)2


Meaning:

average of squared prediction errors

Lower loss = better model.

4. How the Model Learns

The model adjusts:

weight
bias

to reduce the loss.

This process is called:

Gradient Descent

Conceptually:

1. Start with random weight and bias
2. Make predictions
3. Calculate error
4. Adjust weight and bias
5. Repeat until error is small

This is how almost all ML models learn.

5. Visual Intuition

The model is trying to find the best line that fits the data.

Bad line:

y = 0.01x + 0

Better line:

y = 0.04x + 10

Training gradually moves toward the best line.

Day 3 Coding Task (Simple Version)

You will implement the prediction part of linear regression.

Dataset:

import numpy as np

X = np.array([1,2,3,4])
y = np.array([7,9,11,13])

This dataset follows the pattern:

y = 2x + 5
Step 1 — Choose weight and bias
w = 1
b = 0
Step 2 — Prediction
y_pred = w * X + b
Step 3 — Calculate error
error = y_pred - y
Step 4 — Calculate loss
loss = np.mean(error**2)
Your Goal Today

Write a small program that prints:

predictions
error
loss
Expected Output Example
Predictions: [...]
Error: [...]
Loss: ...


```python
import numpy as np

X = np.array([1,2,3,4])
y = np.array([7,9,11,13])

w = 2
b = 5

y_pred = w * X + b

error = y_pred - y
loss = np.mean(error**2)

print("Predicted values:", y_pred)
print("Error:", error)
print("Loss:", loss)    
```
